```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

#Cargamos nuestro archivo CSV previamente ya preparado para su analisis, donde lo cargamos con la libreria pandas y mostramos las primeras 5 filas
df = pd.read_csv('Final.csv')

#Eliminamos la categoria de id de las columnas, esta no es necesaria en general solo es 1 categoria y ya se tiene el nombre
#creamos un df de la correlacion unicamente de los valores numericos
df = df.drop(columns=['ID_CATEGORY'])
dfCorr = df.corr(numeric_only=True)

#De nuestro DF principal nos aseguramos que sea de tipo fecha
df['DATE'] = pd.to_datetime(df['DATE'])
df = df[df['BRAND'] == 'VANISH']

#Sacamos el Total de Unidades y Total de venta de todo nuestro Set de Datos para sacar el % de representacion de cada Producto
totalUnitSales = df['TOTAL_UNIT_SALES'].sum()
totalValueSales = df['TOTAL_VALUE_SALES'].sum()

#TOP 3 PRODUCTOS MAS VENDIDOS DE TODO EL SET DE DATOS
#Buscamos dentro del DF principal los productos con los numeros que se muestran, para poder agrupar posteriormente estos por Producto, formato y Color para sumar
#el total de ventas y de unidades vendidas de esos productos en especifico, agregandoles 2 campos sacando la cuenta para mostrar el % del total que estos productos representan
dfTop3ofAllYearAttrFormat = df[df['ITEM_CODE'].isin(['7702626204208BP1','7501058792808BP1','7501058796578BP2'])]
dfTop3ofAllYearAttrFormat = dfTop3ofAllYearAttrFormat.groupby(['ITEM_CODE','FORMAT','ATTR3'])[['TOTAL_UNIT_SALES','TOTAL_VALUE_SALES']].sum().reset_index()
dfTop3ofAllYearAttrFormat['PERCENTAGE_UNITS'] = ((dfTop3ofAllYearAttrFormat['TOTAL_UNIT_SALES'] * 100) / totalUnitSales).round(2)
dfTop3ofAllYearAttrFormat['PERCENTAGE_SALES'] = ((dfTop3ofAllYearAttrFormat['TOTAL_VALUE_SALES'] * 100) / totalValueSales).round(2)
dfTop3ofAllYearAttrFormat.sort_values(by='TOTAL_UNIT_SALES', ascending= False)

#TOP PRODUCTO MAS VENDIDO DE CADA AÑO, FORMATO Y COLOR
#Sacamos un df agrupado por año, formato, los 3 attrs y el codigo del producto para sumar su Total de Ventas y de Unidades Vendidas ordenado por orden ascendiente
#de año, formato atributos y descendente el valor para solo extraer el mas vendido de esas caracteristicas
dfFormatSalesAtts = (df.groupby(['YEAR','FORMAT','ATTR1','ATTR2','ATTR3','ITEM_CODE'])[['TOTAL_UNIT_SALES','TOTAL_VALUE_SALES']].sum().reset_index()
    .sort_values(by=['YEAR','FORMAT','ATTR1','ATTR2','ATTR3','TOTAL_VALUE_SALES'], ascending = [True,True,True,True,True,False])
    .groupby(['YEAR','FORMAT','ATTR3']).head(1))

#TOP 3 Productos por Formato y Color con MAS UNIDADES Vendidas
top3ProducsSalesUnits = (df.groupby(['ITEM_CODE','FORMAT','ATTR3'])[['TOTAL_UNIT_SALES','TOTAL_VALUE_SALES']].sum().reset_index()
    .sort_values(by='TOTAL_UNIT_SALES',ascending = False))

#TOP 3 Productos por Formato y Color con mayor TOTAL DE VENTAS
top3ProducsSalesValue = (df.groupby(['ITEM_CODE','FORMAT','ATTR3'])[['TOTAL_UNIT_SALES','TOTAL_VALUE_SALES']].sum().reset_index()
    .sort_values(by='TOTAL_VALUE_SALES',ascending = False))
```
